```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.cm as cm
import matplotlib
```

```python
Pr_range = np.logspace(np.log10(0.0001), np.log10(0.9), num=100)
Ts = [0.10, 0.15, 0.20, 0.25]
fig = plt.figure(figsize=(12,10))
cmap = cm.get_cmap('viridis')
color = cmap(1 / 4)         

# Loop through each file
for i, T in enumerate(Ts):
    data= np.zeros((len(Pr_range)))
    for Pr_index, Pr in enumerate(Pr_range):
        # Load the data from the file
        file_path = "./statistical_parameters/Asymp_statistical_parameters_Pr_{:.5f}-T_{:.2f}.txt".format(round(Pr, 5), round(T, 2))
        with open(file_path, "r") as f:
            lines = f.readlines()
        last_value = float(lines[-1].split()[1])/100
        # Store the last value in the data matrix
        data[Pr_index] = last_value
    plt.semilogx(Pr_range, data, linestyle = "-", color = cmap(i / 5), marker = ".", markersize = 10, label=f"$T={T:.2f}$")
    
plt.xlabel("$P_r$", size = 32)
plt.ylabel(r"$\langle S_{central} \rangle$", size=24)
plt.ylim(-0.01,1.05)
ax = plt.gca()
ax.tick_params(axis = 'both', which = 'major', labelsize = 24)
ax.tick_params(axis = 'both', which = 'minor', labelsize = 24)
ax.grid()
plt.legend(fontsize=20)
plt.title(r"$ E=0.1$", fontsize=32)
plt.savefig("./images/S_central.eps", format='eps', bbox_inches='tight', pad_inches=0.1)
plt.show()
plt.close()
```
